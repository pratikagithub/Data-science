```python
import pandas as pd
import numpy as np
df=pd.read_csv('C:\\Users\\HP\\Downloads\\bollywood.csv')
df.head()
```

```python
df.shape
```

```python
df.columns
```

```python
df['MovieName'].sort_values(ascending=True).head(10)
```

```python
df['Genre'].value_counts()
```

```python
df['YoutubeViews'].sort_values(ascending=False).head(10)
```

```python
#nunique() method: returns the number of unique values for each column.
print(df['Genre'].nunique())
```

```python
#filtering
r1=df[df['Genre']=='Romance']
r1[['MovieName']].head()
```

```python
r2=df[(df['Genre']=='Comedy')&(df['YoutubeLikes']>500)]
r2[['MovieName']].head()
```

```python
#query() method
r3=df.query('YoutubeViews>50000')
r3[['MovieName']].head()
```

```python
#group by
r4=df.groupby('Genre')['BoxOfficeCollection'].mean()
r4.sort_values(ascending=True).head()
```

```python
#pivot table
pt1=pd.pivot_table(data=df,index=['Genre'],values=['BoxOfficeCollection','YoutubeViews','YoutubeLikes','YoutubeDislikes'])
print(pt1.shape)
pt1.head()
```

```python
#apply
df[['BoxOfficeCollection','YoutubeViews','YoutubeLikes','YoutubeDislikes']].apply(np.max)
```

```python
#where
res=df['BoxOfficeCollection'].where(df['BoxOfficeCollection']<df['BoxOfficeCollection'].mean(),'High')
res
```

```python
#difference between max and min values
r5=df[['YoutubeLikes']].apply(lambda x:x.max()-x.min())
print(r5)
```

```python
r5=df[['YoutubeDislikes']].apply(lambda x:x.max()-x.min())
print(r5)
```

```python
#assign()
df1=df.assign(Revised_YoutubeViews=lambda x: df['YoutubeViews']+0.50*df['YoutubeViews'])
df1.head()
```

```python
df3=df.assign(RevisedYoutubeLikes=lambda x:1.5*df['YoutubeLikes'],RevisedYoutubeDislikes=lambda x: 1.5*df['YoutubeDislikes'])
df3.head()
```
